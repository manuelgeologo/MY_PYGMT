```python
#primero se importa el modulo pygmt
import pygmt

#creamos una instancia de la clase pygmt.Figure (un objeto)
fig= pygmt.Figure()

#utilizamos el mtodo .coast y sus argumentos para refinir la regin, la proyeccin y plotear la costa.
fig.coast(shorelines=True, region=[-90, -70, 0, 20], projection="M8i", frame=True)

#mostramos la figura
fig.show()
```

```python
#guardar la figura en tu carpeta de trabajo
fig.savefig("america_central.png")
```

```python
# el argumento 'borders' especifica el nivel de los bordes polticos ploteados


#    1 = National boundaries

#    2 = State boundaries within the Americas

#    3 = Marine boundaries

#    a = All boundaries (1-3)
```

```python
#si queremos dibujar bordes polticos con linea negra de grosor 1 pixel. se usa
#bordes="1/1p,black" , pero tambin se pueden dibujar multiples bordes ingresando una lista en vez de un string
```

```python
import pygmt
fig = pygmt.Figure()
# Make a Sinusoidal projection map of the Americas with automatic ticks
#ms sobre proyecciones puede verse en https://desktop.arcgis.com/es/arcmap/10.3/guide-books/map-projections/sinusoidal.htm
#en frame, 'a' pone la grilla automtica, 'f' pone el grosor y 'g' la grilla interior.
fig.basemap(region=[-150, -30, -60, 60], projection="I-90/8i", frame="afg")

# Plot each level of the boundaries dataset with a different color. Ac se da un string para los bordes..
fig.coast(borders=["1/0.5p,black", "2/0.5p,red", "3/0.5p,blue"], land="gray")
fig.show()
```

```python
print ("Ejemplo 3 Terreno con color y agua")
```

```python
#Los parmetros 'land' y 'water' del metodo .coast le asignan un color al relleno de las masas de tierra y de agua.
#Se puede usar los colores por defecto de GMT o un valor numrico de 6 digitos como #333333
```

```python
import pygmt
fig= pygmt.Figure()
#creamos el mapa con proyeccin Mollweide (w) con tamao 8 inches, con ajuste automtico de cuadro (frame)
fig.basemap(region='g',projection='W8i', frame=True)
#creamos los bordes con los colores de agua y tierra que queramos,colores GMT en el link :
from IPython.core.display import display, HTML
display(HTML("""<a href="https://docs.generic-mapping-tools.org/latest/gmtcolors.html">Colores GMT</a>"""))

fig.coast(land='beige', water='lightskyblue')
fig.show()          
```
